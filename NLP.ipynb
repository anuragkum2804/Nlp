```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"E:\Users\User\Files\Spam Email raw text for NLP.csv")
```

```python
df.head()
```

```python
import string
```

```python
df['MESSAGE'] = df['MESSAGE'].str.replace('[{}]'.format(string.punctuation), '')
```

```python
df.drop('FILE_NAME',inplace=True,axis=1)
```

```python
from nltk.corpus import stopwords
remove_words = stopwords.words('english')
pat = r'\b(?:{})\b'.format('|'.join(remove_words))
df['MESSAGE'] = df['MESSAGE'].str.lower()
df['MESSAGE'] = df['MESSAGE'].str.replace(pat, '')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
```

```python
vectorization = CountVectorizer()
X = vectorization.fit_transform(df['MESSAGE'])
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer().fit(X)
```

```python
X_tfidf = tfidf_transformer.transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, df['CATEGORY'], test_size=0.30, random_state = 50)
```

```python
clf = clf.fit(X_train,y_train)
predict = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print (classification_report(y_test, predict))
```

```python
from sklearn import svm
clf = svm.SVC().fit(X_train,y_train)
predict = clf.predict(X_test)
```

```python
print(classification_report(y_test, predict))
```

```python
from catboost import CatBoostClassifier
clf = CatBoostClassifier().fit(X_train,y_train)
predict = clf.predict(X_test)
```

```python
print(classification_report(y_test, predict))
```
